## Spliting the column Memory to two new Columns SSD and HDD.

```python
Data.Memory.value_counts()
```

```python
Data.Memory.size
```

```python
HDD1 = []
SSD1 = []
```

```python
for i in Data.Memory:
    hdd = float('Nan')
    ssd = float('Nan')
    if '+' in i:
        hdd = i.split('+')[1].strip()
        ssd = i.split('+')[0].strip()
        
    elif 'Flash Storage' in i:
        ssd = i
    elif 'Hybrid' in i:
        hdd = i
    else:
        if 'SSD' in i:
            ssd = i
        elif 'HDD' in i:
            hdd = i
    HDD1.append(hdd)
    SSD1.append(ssd)
```
